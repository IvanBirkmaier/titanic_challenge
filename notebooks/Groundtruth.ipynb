### Modify Ground Truth

Modify Gruond_Truth delete unusable rows

```python
ground_truth.drop(['boat', 'body', 'home.dest'], axis=1, inplace=True)
ground_truth = ground_truth.sort_values(by='name')
ground_truth.rename(columns={'name': 'Name'}, inplace=True)
ground_truth.head()
```

```python
# Laden des Datensatz
Data = pd.read_csv(data_path/'../csv/pre_processed_data/train_test_with_sex.csv')
Data = Data.sort_values(by='Name')
Data.head()
```

```python
Data.info()
```

Schauen wo sich die Namen unterscheidnen

```python
# Angenommen, ground_truth und Data sind bereits definierte DataFrames

# Stelle sicher, dass die Namen-Spalten in beiden DataFrames übereinstimmen
ground_truth.rename(columns={'name': 'Name'}, inplace=True)
Data.rename(columns={'name': 'Name'}, inplace=True)

# Führe einen Merge durch und markiere die Herkunft der Daten
merged_df = pd.merge(ground_truth[['Name']], Data[['Name']], on='Name', how='outer', indicator=True)

# Filtere Namen, die nur in einem der beiden DataFrames vorkommen
unique_names_df = merged_df[merged_df['_merge'] != 'both'].copy()

# Ersetze die '_merge'-Spaltenwerte durch eine verständlichere Markierung
unique_names_df['_merge'] = unique_names_df['_merge'].map({'left_only': 'ground_truth', 'right_only': 'Data'})

# Benenne die Spalte für Klarheit um
unique_names_df.rename(columns={'_merge': 'DataFrame'}, inplace=True)

# Überprüfe das Ergebnis
unique_names_df
```

```python
change_names = unique_names_df[unique_names_df["DataFrame"]=="ground_truth"]
change_names
```

Man sieht einfach immer das " nach hinten genommen

```python
# Funktion, die das erste Anführungszeichen entfernt und am Ende einfügt
def adjust_quotes(name):
    # Stelle sicher, dass `name` ein String ist
    name = str(name)
    if '"' in name:
        parts = name.split('"', 1)  # Teilt den String beim ersten Anführungszeichen
        return parts[0] + parts[1] + '"'  # Entfernt das erste Anführungszeichen und fügt es am Ende ein
    return name

# Sicherstellen, dass der Name in beiden DataFrames als String behandelt wird
ground_truth['Name'] = ground_truth['Name'].astype(str)
change_names['Name'] = change_names['Name'].astype(str)

# Erstelle eine Kopie von ground_truth, um die Änderungen vorzunehmen
updated_ground_truth = ground_truth.copy()

# Iteriere über die Namen in change_names
for original_name in change_names['Name']:
    # Wende die adjust_quotes Funktion auf jeden Namen an
    adjusted_name = adjust_quotes(original_name)
    
    # Aktualisiere den Namen in updated_ground_truth, falls vorhanden
    updated_ground_truth.loc[updated_ground_truth['Name'] == original_name, 'Name'] = adjusted_name

# An diesem Punkt enthält updated_ground_truth die aktualisierten Namen
```

Merge:

```python
# Merge durchführen
merged_df = pd.merge(updated_ground_truth, Data[['PassengerId', 'Dataset', 'Name']], on='Name', how='left')
merged_df.head()
```
